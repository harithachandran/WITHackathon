```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_app_ui():
    
    fig, ax = plt.subplots(figsize=(8, 6))
    
    ax.set_facecolor('lightgray')
    
    # Title
    ax.add_patch(patches.Rectangle((0, 0.9), 1, 0.1, linewidth=2, edgecolor='black', facecolor='steelblue'))
    ax.text(0.5, 0.95, 'Need help', ha='center', va='center', fontsize=16, color='white')

    ax.text(0.25, 0.85, 'Your issue', ha='center', va='center', fontsize=12, color='black')
    ax.text(0.25, 0.75, 'Issue description: Fix a broken lock', ha='center', va='center', fontsize=10, color='black')
    ax.text(0.25, 0.65, 'Skills Needed: lock repair', ha='center', va='center', fontsize=10, color='black')

    ax.text(0.75, 0.85, 'Available Female Workers', ha='center', va='center', fontsize=12, color='black')
    ax.text(0.75, 0.75, '1. Alice - lock repair', ha='center', va='center', fontsize=10, color='black')
    ax.text(0.75, 0.65, '2. Maria - lock repair', ha='center', va='center', fontsize=10, color='black')
    ax.text(0.75, 0.55, '3. Emma - lock repair', ha='center', va='center', fontsize=10, color='black')

    # Footer Section 
    ax.add_patch(patches.Rectangle((0, 0), 1, 0.1, linewidth=2, edgecolor='black', facecolor='darkblue'))
    ax.text(0.5, 0.05, 'Hire Worker', ha='center', va='center', fontsize=14, color='white')

    # Remove axis to focus on layout
    ax.axis('off')

    plt.show()

draw_app_ui()
```

```python
from IPython.display import display, HTML  

def notify_verification(success):
    if success:
        message = """
        <div style="
            padding: 15px; 
            background-color: lightgreen; 
            color: black; 
            font-size: 18px; 
            font-weight: bold; 
            border-radius: 5px;
            border: 2px solid green;
            text-align: center;
            margin-top: 10px;
            width: 45%;
            display: inline-block;
            margin-right: 5%;
        ">
            ✅ Facial Recognition Successful! Female worker verified.
        </div>
        """
    else:
        message = """
        <div style="
            padding: 15px; 
            background-color: lightcoral; 
            color: white; 
            font-size: 18px; 
            font-weight: bold; 
            border-radius: 5px;
            border: 2px solid darkred;
            text-align: center;
            margin-top: 10px;
            width: 45%;
            display: inline-block;
        ">
            ❌ Facial Recognition Failed! Worker verification unsuccessful.
        </div>
        """
    return message

# Display both messages
display(HTML(notify_verification(True) + notify_verification(False)))
```
